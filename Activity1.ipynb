<h2>QUESTION 9</h2>
<HR>

<H4>Which year has the highest average flipper length?</H4>

```python
years = penguins["year"].unique()
highest_year = None
highest_avg = 0

for y in years:
    avg_flipper = penguins[penguins["year"] == y]["flipper_length_mm"].mean()
    if avg_flipper > highest_avg:
        highest_avg = avg_flipper
        highest_year = y

highest_year, highest_avg
```

<H2>INSIGHT 9</H2>
<HR>
<H4>The year with the highest average flipper length is 2008, with an average of 202.80 mm.</H4>

<H2>QUESTION 11</H2>
<HR>

<H4>Which species has the widest range in body mass (max - min)?</H4>

```python
species = penguins["species"].unique()
widest_species = None
widest_range = 0

for s in species:
    data = penguins[penguins["species"] == s]["body_mass_g"]
    mass_range = data.max() - data.min()
    if mass_range > widest_range:
        widest_range = mass_range
        widest_species = s

widest_species, widest_range
```

<H2>INSIGHT 11</H2>
<HR>
<H4>The species with the widest range in body mass is Gentoo, with a range of 2350 g.</H4>

<h2>QUESION 13</h2>
<HR>

<h4>Which island has the highest average penguin body mass?</h4>

```python
islands = penguins["island"].unique()
heaviest_island = None
highest_avg_mass = 0

for i in islands:
    avg_mass = penguins[penguins["island"] == i]["body_mass_g"].mean()
    if avg_mass > highest_avg_mass:
        highest_avg_mass = avg_mass
        heaviest_island = i

heaviest_island, highest_avg_mass
```

<h2>INSIGHT 13</h2>
<HR>
<H4>The island with the highest average penguin body mass is Biscoe, with an average of 4712.96 g.</H4>
